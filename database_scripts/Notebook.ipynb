```sql
-- Number of events in the data store
SELECT name, rowcnt,* FROM sys.sysindexes WHERE id = OBJECT_ID('dbo.telemetry')
```

```sql
-- Latency of data ingested
SELECT TOP 1 * FROM dbo.telemetry ORDER BY timestamp DESC
```

```sql
-- Typical time series query
SELECT 
	DATEADD(MINUTE, DATEDIFF(MINUTE, '2000', timestamp) / 10 * 10, '2000') as timeslot,
	deviceid,
	AVG(CONVERT(decimal(4,2),temp)) as avg_temp,
    MIN(CONVERT(decimal(4,2),temp)) as min_temp,
    MAX(CONVERT(decimal(4,2),temp)) as max_temp,
	AVG(CONVERT(decimal(4,2),hum)) as avg_hum,
    MIN(CONVERT(decimal(4,2),hum)) as min_hum,
    MAX(CONVERT(decimal(4,2),hum)) as max_hum
FROM 
	dbo.telemetry 
WHERE 
	timestamp > DATEADD(HOUR, -12, getdate()) AND deviceid IN ('sim000001','sim000002','sim000004','sim000011')
GROUP BY DATEADD(MINUTE, DATEDIFF(MINUTE, '2000', timestamp) / 10 * 10, '2000'), deviceid 
ORDER BY timeslot, deviceid
```
